# EDA

#### Box Plots of the Data:

```python
# List of time points to plot
time_points = ['Time_00', 'Time_01', 'Time_02', 'Time_03', 'Time_04', 'Time_05', 'Time_06', 'Time_07', 'Time_08', 'Time_09', 'Time_10']

# Corresponding labels for the time points
time_labels = [str(i * 10) for i in range(len(time_points))]

# Create a unified DataFrame for plotting
combined_data = pd.DataFrame()

for time_point in time_points:
    if time_point in csv_data:
        # Add a column to identify the time point
        temp_df = csv_data[time_point].copy()
        temp_df['Time_Point'] = time_point
        combined_data = pd.concat([combined_data, temp_df], axis=0)

# Reset the index to avoid duplicate labels
combined_data.reset_index(drop=True, inplace=True)

# Create a box plot for each muscle joint
for muscle in column_headers:
    plt.figure(figsize=(20, 6))
    sns.boxplot(x='Time_Point', y=muscle, data=combined_data, order=time_points)
    plt.title(f'Box Plot for {muscle}')
    plt.xlabel('Channel Units')
    plt.ylabel('Percentage [%]')
    plt.xticks(ticks=range(len(time_points)), labels=time_labels)
    plt.show()
```

#### Line Graph of the Data:

```python
# Define the columns for left and right
left_columns = [col for col in column_headers if 'LT' in col]
right_columns = [col for col in column_headers if 'RT' in col]

# Define the columns for the cervical, upper trap, lumbar, and middle trap
cervical_columns = [col for col in column_headers if 'CERVICAL' in col]
upper_trap_columns = [col for col in column_headers if 'UPPER TRAP' in col]
lumbar_columns = [col for col in column_headers if 'LUMBAR' in col]
middle_trap_columns = [col for col in column_headers if 'MIDDLE TRAP' in col]
```

```python
# Create a figure and axes
fig, axes = plt.subplots(len(left_columns), 1, figsize=(10, 20))

# Iterate over each pair of left and right columns
for i, (left_col, right_col) in enumerate(zip(left_columns, right_columns)):
    # Combine all time CSV values for the current left and right columns
    left_values = pd.concat([csv_data[time][left_col] for time in file_paths.keys() if 'Time' in time])
    right_values = pd.concat([csv_data[time][right_col] for time in file_paths.keys() if 'Time' in time])
    
    # Plot the values
    axes[i].plot(left_values.values, label=left_col)
    axes[i].plot(right_values.values, label=right_col)
    
    # Set the title and labels
    axes[i].set_title(f'{left_col} vs {right_col}')
    axes[i].set_ylabel('Percentage [%]')
    axes[i].legend()

# Set the xlabel for the last subplot
axes[-1].set_xlabel('Time')

# Adjust layout
plt.tight_layout()
plt.show()
```

#### Look at the correlations for each subject at the different time points:

```python
# Calculate the correlation matrix for each time point
correlation_matrices = {time_point: csv_data[time_point].corr() for time_point in time_points}

# Plot the correlation matrices as heatmaps
for time_point, corr_matrix in correlation_matrices.items():
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
    plt.title(f'Correlation Matrix for {time_labels[time_points.index(time_point)]} Channel Units')
    plt.show()

# Output all correlations and from which time point they came if >0.70 or <-0.70 and not 1
correlation_data = []

for time_point, corr_matrix in correlation_matrices.items():
    for i, row in corr_matrix.iterrows():
        for j, value in row.items():
            if (value > 0.7 and value != 1) or (value < -0.7 and value != -1):
                correlation_data.append({'Time_Point': time_point, 'Variable_1': i, 'Variable_2': j, 'Correlation': value})

# Create a DataFrame from the correlation data
correlation_df = pd.DataFrame(correlation_data)
print(correlation_df)
```

```python
# Iterate over each time point and plot the mean muscle activity
for time_point in time_points:
    # Filter the data for the current time point
    time_data = combined_data[combined_data['Time_Point'] == time_point]
    
    # Calculate the mean for each muscle
    time_means = time_data[column_headers].mean()
    
    # Plot the means for each muscle
    plt.figure(figsize=(10, 6))
    time_means.plot(kind='line', marker='o')
    plt.title(f'Average Muscle Activity for {time_labels[time_points.index(time_point)]} Channel Units')
    plt.xlabel('Muscle')
    plt.ylabel('Average Percentage [%]')
    plt.xticks(rotation=45)
    plt.show()
```
