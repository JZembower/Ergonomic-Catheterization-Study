```python
survey_data[radiation_columns].describe()
```

```python
for i, column in enumerate(radiation_columns):
    plt.figure(figsize=(10, 6))
    sns.boxplot(y=survey_data[column])
    plt.title(f'Box Plot of {column}')
    plt.ylabel('Radiation Values')
    plt.xlabel(column)
    if i == 1:  # Change ylim for the second graph
        plt.ylim(0, 5)
    plt.show()
```

```python
# Correlation matrix for radiation columns
corr_radiation = survey_data[radiation_columns].corr()
sns.heatmap(corr_radiation, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
corr_radiation
```

Think about the radiation data now for rampart versus lead individuals.

```python
# Add a new column to indicate the operator type
survey_data['Operator Type'] = survey_data['Participant ID'].apply(lambda x: 'Lead' if x in Lead_Operators else 'Rampart' if x in Rampart_Operators else 'Baseline')

# Filter the data for Lead and Rampart operators
filtered_data = survey_data[survey_data['Operator Type'].isin(['Lead', 'Rampart'])]

# Create box plots for each radiation column with different colors for Lead and Rampart
for i, column in enumerate(radiation_columns):
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Operator Type', y=column, data=filtered_data, palette={'Lead': 'cyan', 'Rampart': 'green'})
    plt.title(f'Box Plot of {column} for Rampart vs Lead Operators')
    plt.ylabel('Radiation Values')
    plt.xlabel('Operator Type')
    if i == 1:  # Change ylim for the second graph
        plt.ylim(0, 5)
    plt.show()
```

```python
# Replace 'Checked' with 1 and 'Unchecked' with 0 in the discomfort data
discomfort_data.replace({'Checked': 1, 'Unchecked': 0}, inplace=True)
```

```python
discomfort_data
```

```python
# Filter discomfort data for Lead and Rampart operators
lead_discomfort_data = discomfort_data[discomfort_data['Participant ID'].isin(Lead_Operators)]
rampart_discomfort_data = discomfort_data[discomfort_data['Participant ID'].isin(Rampart_Operators)]
```

```python
# Count the number of 1's for each column in lead and rampart discomfort data
lead_discomfort_counts = lead_discomfort_data[discomfort_columns].sum()
rampart_discomfort_counts = rampart_discomfort_data[discomfort_columns].sum()
```

Showcase the columns that don't have 0's in them. 

```python
# Filter out columns with all zero values
lead_discomfort_counts_nonzero = lead_discomfort_counts[lead_discomfort_counts != 0]
rampart_discomfort_counts_nonzero = rampart_discomfort_counts[rampart_discomfort_counts != 0]

# Plot the discomfort counts for Lead and Rampart operators
plt.figure(figsize=(14, 7))
lead_discomfort_counts_nonzero.plot(kind='bar', color='cyan', alpha=0.7, label='Lead', position=0)
rampart_discomfort_counts_nonzero.plot(kind='bar', color='green', alpha=0.7, label='Rampart', position=1)
plt.title('Discomfort Counts for Lead and Rampart Operators')
plt.xlabel('Discomfort Categories')
plt.ylabel('Counts')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# output a dataframe table with the discomfort counts above
discomfort_counts = pd.DataFrame({'Lead': lead_discomfort_counts_nonzero, 'Rampart': rampart_discomfort_counts_nonzero})
discomfort_counts
```
