```python
# Define ergonomic thresholds
threshold_min = -20 # degrees
threshold_max = 20 # degrees

# Function to calculate threshold violations
def calculate_threshold_violations(max_data, min_data):
    violations = {}
    for joint in max_data.columns[1:]:  # Exclude 'Unnamed: 0' or case labels
        # Calculate percentage of cases exceeding the thresholds
        above_max = (max_data[joint] > threshold_max).mean() * 100
        below_min = (min_data[joint] < threshold_min).mean() * 100
        violations[joint] = {"above_max": above_max, "below_min": below_min}
    return pd.DataFrame(violations).T

# Calculate threshold violations for 'Lead' and 'Rampart'
lead_violations = calculate_threshold_violations(
    csv_data["lead_max_cases"], csv_data["lead_min_cases"]
)
rampart_violations = calculate_threshold_violations(
    csv_data["rampart_max_cases"], csv_data["rampart_min_cases"]
)

# Prepare the data for visualization
lead_violations["environment"] = "Lead"
rampart_violations["environment"] = "Rampart"
violation_data = pd.concat([lead_violations, rampart_violations])

violation_data
```

- Low Risk Areas:
    - Lumbar Flexion Lateral Movements: No violations were observed in lateral flexion for both left and right sides.
